```python
#임시방편: json->csv로 바꿔 출력하기
import csv
#closeChart_s
#NIChart_s
#stockNumChart_s


def make_csv(inputJson, filename):
    frame='{}.csv'.format(filename)
    with open(frame,'w') as file:
        csv_file=csv.writer(file)
        csv_file.writerow(['period', 'name', 'value'])
        for period in inputJson:
            name=list(inputJson[period].keys())[0]
            value=list(inputJson[period].values())[0]
            print([period,name,value])
            csv_file.writerow([period,name,value])
    return "complete"
        
def make_csv_forClose(inputJson, filename):
    frame='{}.csv'.format(filename)
    with open(frame,'w') as file:
        csv_file=csv.writer(file)
        csv_file.writerow(['date', 'close'])
        for item in inputJson:
            date=item['date']
            close=item['close']
            csv_file.writerow([date,close])
    return "complete"
        
make_csv(NIChart_s,'NIChart_005930')
make_csv(stockNumChart_s,'stockNumChart_005930')
make_csv_forClose(closeChart_s, 'closeChart_005930')
```
