```python
### 에러모음 ###
#회사코드바뀔시 NI 크롤링 불가
#rcpNo 추출시 보고서명이 다른 보고서들 있음()
#에러처리 필요




def close_echo():
    print("close")

def get_dailyClose(tr):
    tds=tr.findAll("td")
    date=tds[0].text
    close=tds[1].text
    return {"date":date, "close":close}

#[{"date":19****,"close":**,***원},{},...] 들의 리스트
#리스트 json파일 저장
def get_pageData(url):
    html=requests.get(url)
    bsObj=BeautifulSoup(html.content,"html.parser")
    table=bsObj.find("table",{"class":"type2"})
    trs=table.findAll("tr")
    pageData=[]
    for tr in trs:
        try : 
            pageData.append( get_dailyClose(tr))
        except Exception as e:pass
    return pageData

def get_closeData(code,page):
    url="https://finance.naver.com/item/sise_day.nhn?code={}&page={}".format(code,page)
    pageData=get_pageData(url)
    return pageData

def get_closeChart(code):
    priceData=[]
    for i in range(1,201):
        data=get_closeData(code,i)
        priceData+=data
    print(priceData)
    return priceData

def make_closeChart(code):
    closeChart=get_closeChart(code)
    print(closeChart)
    fileName="./closeChart_"+code+".json"
    file=open(fileName,"w+")
    file.write(json.dumps(closeChart))
    return []

#get_closeChart("035720")
```

```python
#input= code

# code -> { reportName : [rcpNo, dcmNo], ... }
# put [rcpNo, dcmNo] -> url : http://dart.fss.or.kr/report/viewer.do?rcpNo=20181114001530&dcmNo=6382016&eleId=12&offset=572847&length=21503&dtd=dart3.xsd
# crawl-> { 분기명 : NetIncome , ... }

#output= jsonFile : NetIncomeChart_005930.json
# 17.1 17.2 ... 19.1 19.2 19.3 19.4

#code -> url_list -> rName, rN
#rN -> url_report -> dN
#rN,dN ->docPage-> NetIncome



    
#code -> listUrl 소스코드 생성
def get_listUrl(code):
    conditions=conditions_dic()
    formTypeCode="pblntf_detail_ty=A003&pblntf_detail_ty=A001&pblntf_detail_ty=A002" #연간공시
    API_key=conditions['API_key']
    bgnDate=conditions['bgnDate'] #시작일
    endDate=conditions['endDate'] #종료일
    url_list="https://opendart.fss.or.kr/api/list.xml?crtfc_key={ak}&corp_code={cc}&bgn_de={bd}&end_de={ed}&{f}&page_count=100".format(ak=API_key, bd=bgnDate, ed=endDate, f=formTypeCode, cc=code)
    url_list_result=requests.get(url_list).content.decode('utf-8')
    print('url_list: ' +url_list)
    return url_list_result


#listUrl소스코드 -> reportName s -> 알맞은 형식으로 변경(ex 19.1 18.3 ...)
def rNameEditor(url_list_result):
    reportNms=re.findall(r'<report_nm>(.*?)</report_nm>', url_list_result)
    for i in range(0,len(reportNms)):
        rNm=reportNms[i]
        report_year = str(rNm[7:11])
        
        if rNm[0:2] == '분기': 
            if int(rNm[13]) in range(0,6):reportNms[i]=report_year+".1"
            else : reportNms[i]=report_year+".3"
        elif rNm[0:2] =='반기' : reportNms[i]=report_year+".2"
        elif rNm[0:2] =='사업' : reportNms[i]=report_year+".4"
        else : reportNms[i]="error_"+reportNms[i]
###if문 예외처리, 에러조정 필요!!
    return reportNms
    
#listUrl소스코드 ->rcpNos
def get_rcpNos(url_list_result):   #보류..곧 용도& 구조 변경
    recpt_s=re.findall(r'<rcept_no>(.*?)</rcept_no>', url_list_result) #정규표현식 강의(https://youtu.be/1jdG3K2WhV4) 참고
    return recpt_s

#rcpNos, rNm_edited -> dic{ rcpNo : rNm_edited , ... }
def rNdic(url_list_result):
    recpt_s=get_rcpNos(url_list_result)
    reportNms_edited=rNameEditor(url_list_result)
    dict_rNrNm=dict(zip(recpt_s,reportNms_edited)) #추후 repNo로 검색해 rNm 찾을예정 
    return dict_rNrNm 
    
# 1 rcpNo ->해당dcmNo
def get_dcmNo(rcpNo):
    url_report="http://dart.fss.or.kr/dsaf001/main.do?rcpNo="+rcpNo
    url_report_result=requests.get(url_report).content.decode("UTF-8")
    function_s=re.findall(r'viewDoc(.*?);',url_report_result)
    ingradients=re.split("[']", function_s[0])
    return ingradients[3]

# code-> dic
def rNdNdic(code) :
    url_list_result=get_listUrl(code)
    dict_rNrNm =rNdic(url_list_result)
    dicList={}
    for rN in dict_rNrNm:
        rName=dict_rNrNm[rN]
        dN=get_dcmNo(rN)
        dicList[rName]=[rN,dN] ###key를 뭘로 해야 더 좋을까...?
    #print(dicList)
    return dicList 
    
#rNdNdic('035720')


#url -> NI
def get_NI(url):
    html=requests.get(url)
    bsObj=BeautifulSoup(html.content,"html.parser")
    tbody=bsObj.findAll("tbody")
    trs=tbody[1].findAll("tr")
    tds=trs[28].findAll("td")
    name=tds[0].text
    value=tds[1].text
    return {name:value}

#recNo,dcmNo -> NI
def get_NIofQ(rcpNo,dcmNo):
    #url="http://dart.fss.or.kr/dsaf001/main.do?rcpNo="+rcpNo
    url="http://dart.fss.or.kr/report/viewer.do?rcpNo={}&dcmNo={}&eleId=12&offset=572847&length=21503&dtd=dart3.xsd".format(rcpNo, dcmNo)
    NI=get_NI(url)
    #print(url)
    return NI


#get_NIofQ('20181114001530', '6382016')



#code -> rN,dN s->
def get_NIChart(code):
    NIdata={}
    numDic=rNdNdic(code)
    
    for year in range(2010,2021):
        for qu in range(1,5) :
            try:
                keyName=str(year)+'.'+str(qu)
                rcpNo=numDic[keyName][0]
                dcmNo=numDic[keyName][1]
                NIofQ=get_NIofQ(rcpNo,dcmNo)
                NIdata[keyName]=NIofQ
            except Exception as e:pass
    return NIdata
    

def make_NIChart(code):
    NIChart=get_NIChart(code)
    print(NIChart)
    fileName="./NIChart_"+code+".json"
    file=open(fileName,"w+")
    file.write(json.dumps(NIChart))
    return []

    
def print_url_page(code,eleId):  ###eleId 12:요약연결재무제표, 7:주식의 총수 등
    print('''
    eleId 
    12:요약연결재무제표, 
    7:주식의 총수 등
    ''')
    dic_rNdN=rNdNdic(code)
    url_reports_dic={}
    for rName in dic_rNdN:
        try:    
            rcpNo=dic_rNdN[rName][0]
            dcmNo=dic_rNdN[rName][0]
            url="http://dart.fss.or.kr/report/viewer.do?rcpNo={}&dcmNo={}&eleId={}&offset=572847&length=21503&dtd=dart3.xsd".format(rcpNo, dcmNo, eleId)
            url_reports_dic[rName]=url
        except Exception as e:pass
        
    fileName="./url_page_"+code+".json"
    file=open(fileName,"w+")
    file.write(json.dumps(url_reports_dic))
    return url_reports_dic

```

```python
#url->SN
def get_stockNum(url):###제작중
    html=requests.get(url)
    bsObj=BeautifulSoup(html.content,"html.parser")
    tables=bsObj.findAll('table')
    trs=tables[1].findAll('tr')
    tds=trs[9].findAll('td')
    name=tds[0].text
    value=tds[1].text
    return {name:value}

#recNo,dcmNo -> SN
def get_stockNumOfQ(rcpNo,dcmNo):
    #url="http://dart.fss.or.kr/dsaf001/main.do?rcpNo="+rcpNo
    url="http://dart.fss.or.kr/report/viewer.do?rcpNo={}&dcmNo={}&eleId=7&offset=572847&length=21503&dtd=dart3.xsd".format(rcpNo, dcmNo)
    print(url)
    SN=get_stockNum(url)
    #print(url)
    return SN


#code -> rN,dN s->
def get_stockNumChart(code):
    stockNumdata={}
    numDic=rNdNdic(code)
    
    for year in range(2010,2021):
        for qu in range(1,5) :
            try:
                keyName=str(year)+'.'+str(qu)
                rcpNo=numDic[keyName][0]
                dcmNo=numDic[keyName][1]
                stockNumofQ=get_stockNumOfQ(rcpNo,dcmNo)
                stockNumdata[keyName]=stockNumofQ
            except Exception as e:pass
    return stockNumdata
    

def make_stockNumChart(code):
    stockNumChart=get_stockNumChart(code)
    print(stockNumChart)
    fileName="./stockNumChart_"+code+".json"
    file=open(fileName,"w+")
    file.write(json.dumps(stockNumChart))
    return 'completed'

make_stockNumChart('005930')
#get_stockNumOfQ('20191114001273', '6958001')
```
