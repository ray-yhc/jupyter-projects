```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
#  : ( x''(t) = -sin x(t) )
def odefun(t,Z):
    dZdt = np.zeros(2)
    dZdt[0] = Z[1]
    dZdt[1] = -np.sin(Z[0])
    return dZdt

Za0 = np.array([0,1])
Zb0 = np.array([-5,2])
Zc0 = np.array([5,-2])
Dt=(0,10)
t=np.linspace(0,10)

sol_a = solve_ivp(odefun, Dt, Za0 , dense_output=True)
sol_b = solve_ivp(odefun, Dt, Zb0 , dense_output=True)
sol_c = solve_ivp(odefun, Dt, Zc0 , dense_output=True)

plt.figure(figsize=(9, 10))

plt.subplot(311)
plt.plot(t , sol_a.sol(t).T[:,0])
plt.plot(t , sol_b.sol(t).T[:,0])
plt.plot(t , sol_c.sol(t).T[:,0])
plt.legend(['a', 'b', 'c'])
plt.grid(True, linestyle='--')
plt.title('distance')

plt.subplot(312)
plt.plot(t , sol_a.sol(t).T[:,1])
plt.plot(t , sol_b.sol(t).T[:,1])
plt.plot(t , sol_c.sol(t).T[:,1])
plt.legend(['a', 'b', 'c'])
plt.grid(True, linestyle='--')
plt.title('velocity')
```

```python
# meshgrid 
points = np.arange(-5,5,0.01)
xs, ys = np.meshgrid(points, points)

z = np.sqrt(xs**2 + ys **2) #

plt.imshow(z, cmap=plt.cm.gray); plt.colorbar()
```

```python
# quiver 
arange = np.arange(-5,5,0.5)
y1, y2 = np.meshgrid(arange, arange)

# dZdt = [ Dy1 ; Dy2 ]
Dy1 = y2
Dy2 = -np.sin(y1) 

plt.figure(figsize=(9, 10))
plt.quiver(y1, y2, Dy1, Dy2)
```
